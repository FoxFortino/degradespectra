```python
"""Replace Ia-02cx with Iax in the lnw data files. Originals are backed up."""
for SN in Trn:
    with open(SN, "r") as f:
        lines = f.readlines()
    header = lines[0]
    ind = header.find("Ia-02cx")
    if ind != -1:
        lines[0] = header[:ind] + "Iax" + header[ind+len("Ia-02cx"):]
        
        with open(SN, "w") as f:
            f.write("".join(lines))
```
